```python
brand = soup.find_all('div', class_='s-item__wrapper clearfix')
```

```python
len(brand)
```

```python
name=[]
price=[]
logistics=[]
watching=[]
for i in brand:
  name.append(i.find('h3',class_='s-item__title').text.strip())
  price.append(i.find('span', class_ = 's-item__price').text.strip())
  logistics.append(i.find('span',class_ = 's-item__shipping s-item__logisticsCost').text.strip())
  t=i.find('span', class_ = 'NEGATIVE')
  watching.append(t)
```

```python
dataframe = {'name':name, 'rating':price, 'logistics':logistics, 'action':watching}
```
