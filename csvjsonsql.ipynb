The read_csv() function of pandas not only reads the comma seperated value(csv) file but also can read other type of files like .tsv(tab seperated values) by specifying the sep parameter to '\t' which indicates that the columns in the file are seperated by a tab.

You can also read or fetch the data directly from the website or the URL's by importing requests module and then after pulling the requested file from the URL you can use the read_csv() file.

As I can see that the column has the column names but what if we don't have the appropriate column names then we can just use another parameter of read_csv() function i.e, names(third parameter) and mention the list of column names that you want to use for the table columns.

If not used then the first row of the table is automatically selected as a column name which should not happen at any cost.

Also some times the column names or headers are considered as a row at that time we can use header parameter and make the first column as a header.

not only that there is another parameter that helps to set an index column from multiple unique columns to reduce the complexity that is  **index_col** 


There are more parameter for read_csv like use_cols(specify the number of columns to use), squeeze(to reduce or use only one specific column), nrows(to skip the rows or use the rows as required), encoding(To set the encoding type), error_bad_lines(to skip the bad lines.), dtypes(to set the data types of the specific columns), parse_dates(to handle dates), converters, chunksize(to set the size of the chunk)
