```python
import pandas as pd
import numpy as np
```

```python
dict = {
    "First": [100,200,300,400,np.nan],
    "Second":[30,40,np.nan,50,55],
    "Third":[np.nan,40,50,80,98]
}
```

```python
df2 = pd.DataFrame(dict)
```

```python
df2
```

```python
df2.fillna(10)
```

```python
df2['Fourth'] = df2['First'] + df2['Second']
```

```python
df2
```

```python
df3
```
