```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer as vect
from sklearn.linear_model import LogisticRegression as LR
from sklearn.metrics import accuracy_score as ac 
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colabnotebooks/emails.csv')
```

Pre-processing

```python
df
```

```python
df.isnull()
```

as there are a lots of missing values in the dataset.

```python
mail_data = df.where((pd.notnull(df)),'')
```

```python
mail_data.head()
```

```python
# Checking the number of rows and columns in the datafrome
mail_data.shape
```

Label encoding

```python
#label spam mail as 0; ham mail as 1;
mail_data.loc[mail_data['Category']=='spam', 'Category']=0
mail_data.loc[mail_data['Category']=='ham', 'Category']=1
```

ham = 1 spam = 0

```python
# Seperating the data as texts and label
x = mail_data['Message']
print(x)
y = mail_data['Category']
print(y)
```

Splitting the data into train data and test data 

```python
x_train, x_test, y_train, y_test =train_test_split(x, y, test_size=0.2, random_state=3) 
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

feature extraction

```python
# Transform text to featurevector to use as input for logisticregression data
feature_ext = vect(min_df = 1, stop_words='english', lowercase = True)
```

```python
feature_ext
```

```python
x_train_features = feature_ext.fit_transform(x_train)
x_test_features = feature_ext.transform(x_test)
```

```python
# convert y_train and y_test values as integers
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(x_train_features, x_test_features)
```

Training the model

```python
# Logistic regression
model = LR()
```

```python
# Training the logistic regression model with the training data
model.fit(x_train_features, y_train)
```

Evaluation of trained model

```python
# prediction on training data
pred = model.predict(x_train_features)
```

```python
accuracy_train_data = ac(y_train, pred)
```

```python
print('Accuracy on training data : ', accuracy_train_data)
```

```python
pred_test = model.predict(x_test_features)
accuracy_test_data = ac(y_test, pred_test)
print('Accuracy on test data : ', accuracy_test_data)
```

Building the predictive system

```python
input_data = ["WINNER!! As a valued network customer you have been selected to receivea £900 prize reward! To claim call 09061701461. Claim code KL341. Valid 12 hours only."]

# convert test to feature vectors
input_data_features = feature_ext.transform(input_data)
```

```python
# making prediction
prediction = model.predict(input_data_features)
if prediction == 1: 
  print('ham')
else:
  print('spam')
```
