```python
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.feature_selection import chi2,SelectKBest
```

```python
df = pd.read_csv('/content/drive/MyDrive/ColabNotebooks/train.csv')
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop(columns=['Survived'],axis=1),df['Survived'],test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
x_train.isnull().sum()
```

```python
y_train.sample(5)
```

```python
tr1= ColumnTransformer([
    ('impute_age',SimpleImputer(),[2]),
    ('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])],
    remainder='passthrough')
```

```python
tr2= ColumnTransformer([
    ('ohe_embarked',OneHotEncoder(sparse=False,handle_unknown='ignore'),[1,6])],
    remainder='passthrough') 
```

```python
tr3 = ColumnTransformer([
    ('scale',MinMaxScaler(),slice(0,10))
])
```

```python
tr4 = SelectKBest(score_func=chi2,k=8)
```

```python
tr5 = DecisionTreeClassifier()
```

```python
pipe = Pipeline([
    ('tr1',tr1),('tr2',tr2),('tr3',tr3),('tr4',tr4),('tr5',tr5)])
```

```python
pipe.fit(x_train,y_train)
```

```python
pipe.named_steps
```

```python
from sklearn import set_config
set_config(display='diagram')
```

```python
y_pred = pipe.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
y_test.isnull().sum()
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(pipe, x_train, y_train, cv=5, scoring='accuracy').mean()
```

```python
params = {
    'tr5__max_depth':[1,2,3,4,5,None]}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(pipe, params, cv=5, scoring='accuracy')
grid.fit(x_train, y_train)
```

```python
grid.best_score_
```
